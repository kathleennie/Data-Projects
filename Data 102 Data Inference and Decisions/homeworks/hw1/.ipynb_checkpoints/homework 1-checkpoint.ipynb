```python
# 2a) 
ax= sns.histplot(scipy.stats.zscore(police['x']), stat='density', label='police data')
left, right = ax.get_xlim()
x = np.linspace(left, right, 100)
ax.plot(x, norm.pdf(x), 'r', lw=2, label='norm pdf')
ax.legend()
```

```python
# 2b) 
def benjamini_hochberg(p_values, alpha):
    """
    Returns decisions on p-values using Benjamini-Hochberg.
    
    Inputs:
        p_values: array of p-values
        alpha: desired FDR (FDR = E[# false positives / # positives])
    
    Returns:
        decisions: binary array of same length as p-values, where `decisions[i]` is 1
        if `p_values[i]` is deemed significant, and 0 otherwise
    """

    k=0
    sort=np.sort(p_values)
    
    for i in reversed(range(len(p_values))):
        if sort[i]<=(i+1)*(alpha/len(p_values)):
            break;
    
    decisions = p_values <= sort[i]
    
    return decisions
```

```python
p_values = norm.cdf(-police['x'])
results = benjamini_hochberg(p_values, 0.2)
x = np.count_nonzero(results)
print(x, "discoveries")
```

```python
plt.plot(sorted(p_values), label = 'p values')
plt.axvline(x, label = 'BH Threshold', color = 'green')
plt.legend(loc="lower right")
```

```python
# 3a) 

def difference_means(table, binary, numerical):
    # returns difference betweeen 0 - 1
    means_table = table.groupby(binary).mean()[numerical]
    difference = means_table[0] - means_table[1]
    return difference

def simulate_once(table, binary, numerical):
    category = np.array(table[binary].sample(n=len(table), replace=False).values)
    numer = np.array(table[numerical])
    df = pd.DataFrame({binary: category, numerical: numer})
    return difference_means(df, binary, numerical)

def avg_difference_in_means(binary, numerical):
    
    observed = difference_means(adult, binary, numerical)
    
    differences = []
    rep = 2500
    
    for i in np.arange(rep):
        diff = simulate_once(adult, binary, numerical)
        differences.append(diff)
        
    empirical_p = np.count_nonzero(differences <= observed) / rep
        
    return empirical_p
```

```python
# 3b) 
age_hs = avg_difference_in_means('Post HS?','Age')
age_marry = avg_difference_in_means('Ever Married?','Age')
age_gender = avg_difference_in_means('Gender','Age')
age_50k = avg_difference_in_means('50K?','Age')

hours_hs = avg_difference_in_means('Post HS?','Hours per Week')
hours_marry = avg_difference_in_means('Ever Married?','Hours per Week')
hours_gender = avg_difference_in_means('Gender','Hours per Week')
hours_50k = avg_difference_in_means('50K?','Hours per Week')
```

```python
# 3c) 
pv = np.array([age_hs, age_marry, age_gender, age_50k, hours_hs, hours_marry, hours_gender, hours_50k])
pv < 0.05
```

Given the above output, we can reject the following combinations:                                                            (Age, Marriage) (Age, 50k) (Hours per Week, Marriage) (Hours per week, Gender) (Hours per Week, 50k) 

```python
# 3d) 

fwer = 0.05/len(pv)

pv < fwer

# True means reject while False means don't reject
```

Given the output above, we would only reject (Age, Marriage) 

```python
# 3e) 

benjamini_hochberg(pv, 0.05)

# True means reject while False means don't reject
```

Given the above output, we would reject the same groups as we did in 3c) which are: (Age, Marriage) (Age, 50k) (Hours per Week, Marriage) (Hours per week, Gender) (Hours per Week, 50k)

```python
# 3f) 
```

In d) we only rejected one test while in e) we rejected 5 tests. This makes sense sense BH controls primarily for FDR (allows for number of Type I error in prorportion to true positives) while Bonferroni primarily controls the FWER (probability of making Type I error at all). While BH punishes p-values according to their ranking, Bonferroni punishes all p-values equally. As a result, Bonferroni is known to be more conservative than BH which is why we see only 1 rejection in d) but 5 in e).
