### 2d

```python
def Gibbs_sampler(Z, T, alpha, beta, tau):
    temp = Z
    n = X.shape[0]
    m = X.shape[1]
    for t in range(T):
        for i in range(n):
            for j in range(m):
                neighbors = [(i, j-1), (i-1, j), (i+1, j), (i, j+1)]
                sum_neighbors = sum(Z[x, y] for (x, y) in neighbors if 0 <= x < n and 0 <= y < m)
                temp[i, j] = ((tau * X[i, j]) + beta * sum_neighbors) / (alpha + tau) + np.random.randn() * np.sqrt( 1 / (alpha + tau))
    return temp

Z_1 = plt.imshow(Gibbs_sampler(X.copy(), 1, 250, 62.5, 0.01), cmap = 'gray')
```

since at T = 1 it took about 6 seconds, at T = 100 would take about 10 minutes

### 2f


```python
def Blocked_Gibbs_sampler(Z, T, alpha, beta, tau):
    i, j = Z.shape
    
    for t in range(T):
        delta = np.random.randn(i, j) * np.sqrt(1 / (alpha + tau))
        Z_temp = np.pad(Z, 1)
        
        u = Z_temp[2:, 1:-1] + Z_temp[1:-1, 2:] + Z_temp[:-2, 1:-1] + Z_temp[1:-1, :-2]
        Z[::2, 1::2] = (tau / (alpha + tau)) * X[::2, 1::2] + (beta / (alpha + tau)) * u[::2, 1::2] + delta[::2, 1::2]
        Z[1::2, ::2] = (tau / (alpha + tau)) * X[1::2, ::2] + (beta / (alpha + tau)) * u[1::2, ::2] + delta[1::2, ::2]
        
        u2 = Z_temp[2:, 1:-1] + Z_temp[1:-1, 2:] + Z_temp[:-2, 1:-1] + Z_temp[1:-1, :-2]
        Z[::2, ::2] = (tau / (alpha + tau)) * X[::2, ::2] + (beta / (alpha + tau)) * u2[::2, ::2] + delta[::2, ::2]
        Z[1::2, 1::2] = (tau / (alpha + tau)) * X[1::2, 1::2] + (beta / (alpha + tau)) * u2[1::2, 1::2] + delta[1::2, 1::2]
        
    return Z
plt.imshow(Blocked_Gibbs_sampler(X.copy(), 100, 250, 62.5, 0.01), cmap = 'gray')
```

Blocked gibbs took about 11 seconds to run

## Question 3

```python
elections = pd.read_csv('us_elections.csv')
elections.head()
```

### 3a

```python
import pymc3 as pm

with pm.Model() as model:
    priors = {"govtrack_ideology": pm.Normal.dist(mu = 0, sigma = 1), 
              "house_dem18_margin": pm.Normal.dist(mu = 0, sigma = 1)}
    pm.glm.GLM.from_formula('house_dem20_margin ~ 0 + govtrack_ideology + house_dem18_margin', 
                            elections, family = pm.glm.families.Normal(), priors = priors)
    trace = pm.sample(1000)

plt.scatter(trace['govtrack_ideology'], trace['house_dem18_margin'])
plt.xlabel('Current Officeholder Ideology Score')
plt.ylabel('2018 Election Democratic Margin')
plt.title('Sigma = 1')
plt.xlim(-0.36, 0)
plt.ylim(-0.1, 1)
```

```python
with pm.Model() as model:
    priors = {"govtrack_ideology": pm.Normal.dist(mu = 0, sigma = 1), 
              "house_dem18_margin": pm.Normal.dist(mu = 0, sigma = 0.1)}
    pm.glm.GLM.from_formula('house_dem20_margin ~ 0 + govtrack_ideology + house_dem18_margin', 
                            elections, family = pm.glm.families.Normal(), priors = priors)
    trace = pm.sample(1000)

plt.scatter(trace['govtrack_ideology'], trace['house_dem18_margin'])
plt.xlabel('Current Officeholder Ideology Score')
plt.ylabel('2018 Election Democratic Margin')
plt.title('Sigma = 0.1')
plt.xlim(-0.36, 0)
plt.ylim(-0.1, 1)
```

```python
with pm.Model() as model:
    priors = {"govtrack_ideology": pm.Normal.dist(mu = 0, sigma = 1), 
              "house_dem18_margin": pm.Normal.dist(mu = 0, sigma = 0.01)}
    pm.glm.GLM.from_formula('house_dem20_margin ~ 0 + govtrack_ideology + house_dem18_margin', 
                            elections, family = pm.glm.families.Normal(), priors = priors)
    trace = pm.sample(1000)

plt.scatter(trace['govtrack_ideology'], trace['house_dem18_margin'])
plt.xlabel('Current Officeholder Ideology Score')
plt.ylabel('2018 Election Democratic Margin')
plt.title('Sigma = 0.01')
plt.xlim(-0.36, 0)
plt.ylim(-0.1, 1)
```

### 3b

Visually comparing the above plots, we can see that sigma = 1 and sigma = 0.1 are more similar to each other. In contrast, sigma = 0.01 is very different because the significantly smaller value forces our prior to be centered around 0 meaning the predictive coefficient has less space for variation. In comparison, sigma = 1 and sigma = 0.1 have a lot more vertical freedom.

### 3c

When we use a small sigma value, we believe that the current officeholder ideology score has a very small and limited effect to people's decision making process in the election, giving it minimal effect on predicting eletion outcomes. In contrast, a large sigma value means we believe that the current officeholder ideology has greater effect on predicting election results
