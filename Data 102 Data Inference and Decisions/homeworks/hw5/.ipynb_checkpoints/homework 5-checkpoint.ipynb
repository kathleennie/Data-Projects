```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn 
import scipy.stats as stats
```

### Question 2

#### 2a)

```python
A = np.array([1,2])
c=4
k=2
n=50
B = 2000
G_n= np.array([])
G_n2 = np.array([])
```

```python
def explore_commit(A, c, k, n):
    initial= np.repeat(0.0, k) 
    for i in range(c*k):
        reward = np.random.normal(loc= A-1, scale= 1)
        initial += reward
        optimal = np.where(initial==(max(initial)))[0][0] + 1
            
    best = 0 
    for i in range(c*k, n):
        best += np.random.normal(loc= A[optimal - 1] - 1, scale= 1)
    return best
    
```

```python
for i in range(B):
    G_n = np.append(G_n, explore_commit(A, c, k, n))
    
p_regret_etc = np.mean(50*(max(A) - 1) - G_n)
p_regret_etc
```

The average pseudo regret for explore then commit is about 8.55

```python
plt.hist(G_n, density=True, bins = 20)
plt.title("histogram of rewards: explore then commit")
```

#### 2b)

```python
def ucb(A, c, k, n):
    total = np.repeat(0.0, k) 
    for i in range(c*k):
        reward = np.random.normal(loc= A-1, scale= 1)
        total += reward
        
    T_a = np.repeat(float(c), k)
 
    for i in range(c*k, n):
        ucb = total/T_a + np.sqrt(2*np.log(20)/T_a)
        optimal = np.where(ucb == max(ucb))[0][0]
        total[optimal] += np.random.normal(loc= A[optimal] - 1, scale = 1)
        T_a[optimal] += 1
        
    return sum(total)
```

```python
for i in range(B):
    G_n2 = np.append(G_n2, ucb(A, c, k, n))
    
p_regret_UCB = np.mean(50*(max(A) - 1) - G_n2)
p_regret_UCB
```

The pseudo regret for UCB is about 1.33 This significantly lower than our pseudo regret from part a) which was about 8.9 By the pseudo regret metric, this means UCB is performing better than ETC

```python
plt.hist(G_n2, density=True, bins = 20)
plt.title("histogram of rewards: UCB")
```

#### 2c) 


```python
plt.hist(G_n, alpha=0.5, density=True, label='etc', bins = 20)
plt.hist(G_n2, alpha=0.5, density=True, label='ucb', bins = 20)
plt.legend()
plt.title("total rewards comparison")
```

The mean of the UCB distribution seems to be higher than the ETC distribution for total rewards. This difference in distribution is because UCB selects arms based on the highest mean and upper confidence bound, so it tends to select higher in comparison to ETC. ETC on the other had sees two different concentrated values at 0 and around 40. This is because sometimes, ETC chooses worse arm and repeatedly pulls on it, causing the total rewards to be a lot smaller than what is normally seen
