```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn 
```

### Question 1

#### a)

```python
df = pd.read_csv("ihdp.csv")
treatment = df['treatment']
outcome = df['outcome']
features = df.drop(['treatment', 'outcome'], axis = 1)

from sklearn.linear_model import LogisticRegression as LR
lr = LR(penalty='none', max_iter=200, random_state=0)

lr.fit(features, treatment)
```

#### b) 

```python
def estimate_treatment_effect(lr, X, Y, Z):
    predictions = lr.predict_proba(X)[:, 1]
    avg_effect = np.mean(Z*Y/predictions) - np.mean(((1-Z)*Y)/(1-predictions))
    return avg_effect
```

#### c) 


```python
estimate_treatment_effect(lr, features, outcome, treatment)
```

the average treatment effect of the IHDP dataset is about 3.697 which is a positive number, implying that the treatment had a beneficial causal effect on the child cognitive test score outcome

### Question 3

#### c)

```python
delta = np.geomspace(1/2, 1/1000, 10)
log_delta = np.log(delta)
a = 1/(delta * np.power(0.1, 2))
log_a = np.log(a)

plt.plot(log_delta, log_a)
plt.title("Number of Samples Using  Chebyshev's")
```

```python
delta = np.geomspace(1/2, 1/1000, 10)
log_delta = np.log(delta)
b = np.log(2/delta) * (4/ (2*np.power(0.1, 2)))
log_b = np.log(b)

plt.plot(log_delta, log_b)
plt.title("Number of Samples Using Hoeffding’s")
```

As we can see above, the bounds we derived from part a) shows a negative linear relationship between the log delta on the x axis and log n samples on the y axis. On the other hand, the bounds we derived from part b) shows a negative logarithmic function instead. In situations where we can't assume that each X_i is bounded, then our concentration inequalities wouldn't hold because the conditions for Chebyshevs and Hoeffding's assume our variables are bounded and their expected values can be calculated.
