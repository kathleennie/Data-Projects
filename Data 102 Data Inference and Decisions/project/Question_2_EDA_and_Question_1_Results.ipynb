# Data Cleaning

```python
asthma_data_raw  = pd.read_csv("U.S._Chronic_Disease_Indicators__Asthma.csv")
asthma_data1 = asthma_data_raw.copy()
```

```python
#clean the asthma dataset
#the dataset to only include YearStart, Location Abbr, StratificationCategoryID1, StratificationID1, DataValueUnit, DataValueType, and DataValue
asthma_data1 = asthma_data1[['YearStart', 'LocationDesc', 'LocationAbbr', 'DataValueUnit', 'DataValueType', 
'DataValue', 'StratificationCategoryID1', 'StratificationID1']]
#asthma_data1.head()
```

```python
#clean the asthma dataset so that..
##it only includes the rows that show cases per 10,000
##stratification ID1 = Gender
##only shows age adjusted rates
##data value must be a number
asthma_data1 = asthma_data1[asthma_data1['DataValueUnit'] == 'cases per 10,000']
asthma_data1 = asthma_data1[asthma_data1['DataValueType'] == 'Age-adjusted Rate']
asthma_data1 = asthma_data1[asthma_data1['StratificationCategoryID1'] == 'GENDER']
asthma_data1 = asthma_data1[pd.to_numeric(asthma_data1['DataValue'], errors='coerce').notnull()]
#asthma_data1.head()
```

```python
#only include years 2011 to 2014 in Asthma dataset because that is the only data provided in the pm2.5 dataset
asthma_data1 = asthma_data1[(asthma_data1['YearStart'] >= 2011) & (asthma_data1['YearStart'] <= 2014)]
#asthma_data1.head()
```

```python
#clean asthma dataset to only include what is needed
asthma_data1 = asthma_data1.drop(['LocationDesc', 'LocationAbbr', 'DataValueUnit', 'DataValueType', 
'StratificationCategoryID1', "YearStart"], axis = 1)
#asthma_data1.head()
```

```python
#prepare df for merging
#asthma_data1['FIPS_Code'] = asthma_data1['FIPS_Code'].astype(int)
ozone_data = ozone_data.dropna(subset='statefips')
ozone_data['statefips'] = ozone_data['statefips'].astype(int)
asthma_data1 = asthma_data1.rename(columns = {'FIPS_Code': 'statefips'})
```

```python
asthma_data1['statefips'] = asthma_data1['statefips'].astype(int)
```

```python
#Merge the data
merged_df = asthma_data1.merge(ozone_data, how="left", on = "statefips")
merged_df = merged_df[pd.to_numeric(merged_df['ds_o3_pred'], errors='coerce').notnull()]
```

```python
confounders_df = pd.read_csv("/content/ACSDP1Y2014.DP03-2023-05-03T211525.csv")
#confounders_df
```

```python
# Extract rates of interest (2014 data)
# 9 - Percent Unemployed in Civilian Labor Force
# 79 - With Food Stamp/SNAP benefits in the past 12 months rate
# 105 - Civilian Noninstitutionalized rate with no health insurance coverage
# 135 - Rate of all people below poverty level
confounders_df=confounders_df.loc[[9, 79, 105, 135],::2]
```

```python
# Rate of health coverage and Food Stamp benefits doesn't appear to be confounding
unemployment_rates = confounders_df.transpose().iloc[1:, 0].str.rstrip("%").astype('float')
#snap_rate = confounders_df.transpose().iloc[1:, 1].str.rstrip("%").astype('float')
#no_ins_rate = confounders_df.transpose().iloc[1:, 2].str.rstrip("%").astype('float')
poverty_rate = confounders_df.transpose().iloc[1:, 3].str.rstrip("%").astype('float')
```

```python
# Create DataFrame with confounder rates
conf_df = pd.DataFrame({"Unemployment Rate": unemployment_rates,
                        "Poverty Rate": poverty_rate})
conf_df.index.name = "LocationDesc"
conf_df = conf_df.reset_index()
```

```python
# Remove suffixes from state names
conf_df["LocationDesc"] = conf_df["LocationDesc"].str.removesuffix("!!Percent")
# Remove Puerto Rico
conf_df = conf_df.iloc[:51,:]
conf_df['statefips'] = conf_df['LocationDesc'].map(fips_dict)
conf_df['statefips'] = conf_df['statefips'].astype(int)
```

```python
merged_df = merged_df.merge(conf_df, how="left", on = "statefips")
merged_df = merged_df[pd.to_numeric(merged_df['ds_o3_pred'], errors='coerce').notnull()]
```

```python
#convert datasets to csv due to lack of memory
merged_df.to_csv('Asthma and Ozone.csv', index=False)
asthma_data1.to_csv("Asthma Cleaned Data.csv", index = False)
```

# Regression

```python
#read csv
df = pd.read_csv("Asthma and Ozone.csv")
```

```python
#clean up nulls in df
df = df[pd.to_numeric(df['ds_o3_pred'], errors='coerce').notnull()]
```

```python
y = df["DataValue"]
X = df[["ds_o3_pred", "Poverty Rate", "Unemployment Rate"]]
df_1 = df.sample(n=10000)
```

```python
sns.scatterplot(data=df_1, x="Unemployment Rate",y="DataValue")
plt.title("State Unemployment Rate vs Asthma Hospitalizations")
plt.ylabel("Hospitalizations (cases per 10,000)")
```

```python
sns.scatterplot(data=df_1, x="Poverty Rate",y="DataValue")
plt.title("State Poverty Rate vs Asthma Hospitalizations")
plt.ylabel("Hospitalizations (cases per 10,000)")
```

```python
#sns.scatterplot(data=df_1, x="ds_o3_pred",y="DataValue")
sns.regplot(data = df_1, x = 'ds_o3_pred', y = 'DataValue', fit_reg = False,
           y_jitter = 1, scatter_kws = {'alpha' : 0.05})
plt.title("Ozone Concentration vs Asthma Hospitalizations")
plt.ylabel("Hospitalizations (cases per 10,000)")
```

```python
sns.regplot(data = df_1, x = 'ds_o3_pred', y = 'Poverty Rate', fit_reg = False,
           y_jitter = 0.5, scatter_kws = {'alpha' : 0.05})
plt.title("Ozone Concentration vs Poverty Rate")
```

```python
sns.regplot(data = df_1, x = 'ds_o3_pred', y = 'Unemployment Rate', fit_reg = False,
           y_jitter = 0.25, scatter_kws = {'alpha' : 0.1})
plt.title("Ozone Concentration vs State Unemployment Rate")
```

```python
sns.scatterplot(data=df_1, x="Unemployment Rate",y="Poverty Rate")
plt.title("State Unemployment Rate vs Poverty Rate")
```
