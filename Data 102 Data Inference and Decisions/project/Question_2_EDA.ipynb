```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import pymc3 as pm
from pymc3 import glm
import statsmodels.api as sm
import arviz
import seaborn as sns
```

```python
asthma_data = pd.read_csv('U.S._Chronic_Disease_Indicators__Asthma.csv')
```

```python
counts_data = asthma_data[['YearStart', 'LocationDesc', 'LocationAbbr', 'DataValue', 'DataValueType']]
```

```python
counts_data
```

```python
counts_data_cruderate = counts_data[counts_data['DataValueType'] == "Crude Prevalence"]
ax = sns.histplot(data = counts_data_cruderate, x = "DataValue")
ax.set(title = "histogram of Crude Prevalence", xlabel = "DataValue")
plt.show()
```

```python
mean_value = counts_data_cruderate["DataValue"].mean()
print(mean_value)
```

```python
num_rows = counts_data_cruderate.shape[0]
print(num_rows)
```

```python
counts_data_AArate = counts_data[counts_data['DataValueType'] == "Age-adjusted Rate"]
ax = sns.histplot(data = counts_data_AArate, x = "DataValue")
ax.set(title = "histogram of Age-adjusted Rate", xlabel = "DataValue")
plt.show()
```

```python
mean_value = counts_data_AArate["DataValue"].mean()
print(mean_value)
```

```python
num_rows = counts_data_AArate.shape[0]
print(num_rows)
```

```python
sns.histplot(data = counts_data_cruderate, x = "DataValue", label = 'Crude Prevalence', color = 'pink')
sns.histplot(data = counts_data_AArate, x = "DataValue", label = 'Age-adjusted Rate')
plt.legend()
plt.title("Crude Prevalence vs Age-adjusted Rate")
```

```python
year = [2015]
counts_data_2015 = counts_data[counts_data['YearStart'].isin(year)]
counts_data_AArate_2015 = counts_data_2015[counts_data_2015['DataValueType'] == "Age-adjusted Rate"]
sns.histplot(data = counts_data_AArate_2015, x = "DataValue", label = '2015', color = 'yellow')

year = [2018]
counts_data_2018 = counts_data[counts_data['YearStart'].isin(year)]
counts_data_AArate_2018 = counts_data_2018[counts_data_2018['DataValueType'] == "Age-adjusted Rate"]
sns.histplot(data = counts_data_AArate_2018, x = "DataValue", label = '2018')

year = [2016]
counts_data_2016 = counts_data[counts_data['YearStart'].isin(year)]
counts_data_AArate_2016 = counts_data_2016[counts_data_2016['DataValueType'] == "Age-adjusted Rate"]
sns.histplot(data = counts_data_AArate_2016, x = "DataValue", label = '2016', color = 'green')

year = [2017]
counts_data_2017 = counts_data[counts_data['YearStart'].isin(year)]
counts_data_AArate_2017 = counts_data_2017[counts_data_2017['DataValueType'] == "Age-adjusted Rate"]
sns.histplot(data = counts_data_AArate_2017, x = "DataValue", label = '2017', color = 'red')

plt.legend()
plt.title('Histogram of 2015-2018 Age-adjusted Rates')
plt.show()
```

```python
counts_data_cruderate_2015 = counts_data_2015[counts_data_2015['DataValueType'] == "Crude Rate"]
sns.histplot(data = counts_data_cruderate_2015, x = "DataValue", label = '2015', color = 'yellow')

counts_data_cruderate_2018 = counts_data_2018[counts_data_2018['DataValueType'] == "Crude Rate"]
sns.histplot(data = counts_data_cruderate_2018, x = "DataValue", label = '2018')

counts_data_cruderate_2016 = counts_data_2016[counts_data_2016['DataValueType'] == "Crude Rate"]
sns.histplot(data = counts_data_cruderate_2016, x = "DataValue", label = '2016', color = 'green')

counts_data_cruderate_2017 = counts_data_2017[counts_data_2017['DataValueType'] == "Crude Rate"]
sns.histplot(data = counts_data_cruderate_2017, x = "DataValue", label = '2017', color = 'red')

plt.title('Histogram of 2015-2018 Crude Prevalence Rates')
plt.legend()
plt.show()
```
