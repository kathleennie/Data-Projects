## Question 2 Random Forest 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import seaborn as sns
```

```python
asthma = pd.read_csv("asthma.csv")
asthma.head()
```

```python
asthma.groupby(['Question'])['Question'].count()
```

```python
as_filt = asthma[(asthma['YearStart'] > 2012) & (asthma['YearStart'] < 2019) & 
                 (asthma['Question'] == 'Current asthma prevalence among adults aged >= 18 years') & 
                 (asthma['StratificationCategory1'].isin(['Overall', 'Gender']))]
as2 = as_filt[as_filt['DataValue'].notna()]

final = as2[['YearStart','LocationAbbr', 'DataValueType', 'DataValue','StratificationCategory1','Stratification1']]
final
```

```python
state_to_region = state_to_region = {
    'AK': 'West',
    'AL': 'South',
    'AR': 'South',
    'AZ': 'West',
    'CA': 'West',
    'CO': 'West',
    'CT': 'Northeast',
    'DE': 'South',
    'FL': 'South',
    'GA': 'South',
    'HI': 'West',
    'IA': 'Midwest',
    'ID': 'West',
    'IL': 'Midwest',
    'IN': 'Midwest',
    'KS': 'Midwest',
    'KY': 'South',
    'LA': 'South',
    'MA': 'Northeast',
    'MD': 'South',
    'ME': 'Northeast',
    'MI': 'Midwest',
    'MN': 'Midwest',
    'MO': 'Midwest',
    'MS': 'South',
    'MT': 'West',
    'NC': 'South',
    'ND': 'Midwest',
    'NE': 'Midwest',
    'NH': 'Northeast',
    'NJ': 'Northeast',
    'NM': 'West',
    'NV': 'West',
    'NY': 'Northeast',
    'OH': 'Midwest',
    'OK': 'South',
    'OR': 'West',
    'PA': 'Northeast',
    'RI': 'Northeast',
    'SC': 'South',
    'SD': 'Midwest',
    'TN': 'South',
    'TX': 'South',
    'UT': 'West',
    'VA': 'South',
    'VT': 'Northeast',
    'WA': 'West',
    'WI': 'Midwest',
    'WV': 'South',
    'WY': 'West'
}

final['Region']= final['LocationAbbr'].map(state_to_region)
final
```

```python
dummy_df = pd.get_dummies(final['Region'])
dummy_df2 = pd.get_dummies(final['Stratification1'])
final = pd.concat([final, dummy_df, dummy_df2], axis = 1)

final_female = final[final['Female'] == 1]
final_male = final[final['Male'] == 1]
final_overall = final[final['Overall'] == 1 ] 
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model_f = RandomForestRegressor(max_features=1)
forest_model_m = RandomForestRegressor(max_features=1)
forest_model_o = RandomForestRegressor(max_features=1)

from sklearn.model_selection import train_test_split

train_female, test_female = train_test_split(final_female, test_size=0.3, random_state=101)
train_male, test_male = train_test_split(final_male, test_size=0.3, random_state=101)
train_overall, test_overall = train_test_split(final_overall, test_size=0.3, random_state=101)
```

```python
X_cols = ['YearStart', 'Midwest', 'Northeast', 'South', 'West'] 
Y_col = 'DataValue'

X_female = train_female.loc[:, X_cols]
y_female = train_female.loc[:, Y_col]

X_male = train_male.loc[:, X_cols]
y_male = train_male.loc[:, Y_col]

X_overall = train_overall.loc[:, X_cols]
y_overall = train_overall.loc[:, Y_col]

forest_model_f.fit(X_female, y_female)
forest_model_m.fit(X_male, y_male)
forest_model_o.fit(X_overall, y_overall)

train_female["forest_pred"] = forest_model_f.predict(train_female.loc[:, X_cols])
test_female["forest_pred"] = forest_model_f.predict(test_female.loc[:, X_cols])

train_male["forest_pred"] = forest_model_m.predict(train_male.loc[:, X_cols])
test_male["forest_pred"] = forest_model_m.predict(test_male.loc[:, X_cols])

train_overall["forest_pred"] = forest_model_o.predict(train_overall.loc[:, X_cols])
test_overall["forest_pred"] = forest_model_o.predict(test_overall.loc[:, X_cols])
```

```python
trainfemale_rmse = np.mean((train_female["forest_pred"] - train_female["DataValue"]) ** 2) ** 0.5
testfemale_rmse = np.mean((test_female["forest_pred"] - test_female["DataValue"]) ** 2) ** 0.5

trainmale_rmse = np.mean((train_male["forest_pred"] - train_male["DataValue"]) ** 2) ** 0.5
testmale_rmse = np.mean((test_male["forest_pred"] - test_male["DataValue"]) ** 2) ** 0.5

train_rmse = np.mean((train_overall["forest_pred"] - train_overall["DataValue"]) ** 2) ** 0.5
test_rmse = np.mean((test_overall["forest_pred"] - test_overall["DataValue"]) ** 2) ** 0.5

print("Training set error for random forest female:", trainfemale_rmse)
print("Test set error for random forest female: ", testfemale_rmse)

print("Training set error for random forest male:", trainmale_rmse)
print("Test set error for random forest male: ", testmale_rmse)

print("Training set error for random forest overall:", train_rmse)
print("Test set error for random forest overall: ", test_rmse)
```

### create prediction intervals for the overall category 

```python
X_train = train_overall.loc[:, X_cols]
y_train = train_overall.loc[:, Y_col]
X_test = test_overall.loc[:, X_cols]

# Number of bootstrapped models
n_models = 100

# Training multiple RandomForestRegressor models using different bootstrap samples
models = []
for i in range(n_models):
    bootstrap_indices = np.random.choice(np.arange(len(X_train)), size=len(X_train), replace=True)
    X_train_bootstrap = X_train.iloc[bootstrap_indices]
    y_train_bootstrap = y_train.iloc[bootstrap_indices]

    model = RandomForestRegressor(max_features=1, random_state=i)
    model.fit(X_train_bootstrap, y_train_bootstrap)
    models.append(model)

# Make predictions using each model for the new data points in the test set
predictions = []
for model in models:
    preds = model.predict(X_test)
    predictions.append(preds)

# Calculate the percentiles of the predictions to obtain the prediction interval
predictions = np.array(predictions)
lower = np.percentile(predictions, 2.5, axis=0)
upper = np.percentile(predictions, 97.5, axis=0)


# Forest model prediction for train and test set
test_overall["lower prediction"] = lower
test_overall["upper prediction"] = upper
```
