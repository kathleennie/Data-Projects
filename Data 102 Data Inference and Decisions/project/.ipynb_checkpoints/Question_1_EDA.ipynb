## EDA Research Question 1: Estimated Ozone Concentrations vs Latitude/Longitude data 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import seaborn as sns
```

```python
ozone = pd.read_csv("ozone.csv")
ozone = ozone.sample(n= 100000)

ozone2012 = pd.read_csv("ozone2012.csv")
ozone2012 = ozone2012.sample(n = 100000)

ozone2013 = pd.read_csv("ozone2013.csv")
ozone2013 = ozone2013.sample(n = 100000)

ozone2014 = pd.read_csv("ozone2014.csv")
ozone2014 = ozone2014.sample(n = 100000)
ozone2014

ozone_merged = pd.concat([ozone, ozone2012, ozone2013, ozone2014])
ozone_merged = ozone_merged.reset_index(drop=True)
ozone_merged
```

```python
ax = sns.histplot(data = ozone_merged, x = "ds_o3_pred")
ax.set(title = "histogram of estimated ozone concentrations", xlabel = "ozone concentrations")
plt.show()
```

```python
from matplotlib.pyplot import figure

fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
fig.set_size_inches(12, 5)
sns.scatterplot(data=ozone, x= "latitude", y="ds_o3_pred", ax=ax1)
ax1.set(ylabel='Estimated Ozone Concentration')
sns.scatterplot(data=ozone, x= "longitude", y="ds_o3_pred", ax=ax2)
fig.suptitle("ozone 2011 vs geographical data")
fig.show()

```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
fig.set_size_inches(12, 5)
sns.scatterplot(data=ozone2012, x= "latitude", y="ds_o3_pred", ax=ax1)
ax1.set(ylabel='Estimated Ozone Concentration')
sns.scatterplot(data=ozone2012, x= "longitude", y="ds_o3_pred", ax=ax2)
fig.suptitle("ozone 2012 vs geographical data")
fig.show()
```
